```python
from parse import parse

filename = "A2007144191000.L2_LAC_OC.x.nc"

# use the parse library to extract 3 parameters from the given filename
# The parameters:
#    dt : "datetime" a decimal number (d)
#    level : the processing level - a decimal number (d)
#    product_type : a series of letters (l)
params_parsed = parse("A{dt:d}.L{level:d}_LAC_{product_type:l}.x.nc", filename)

print(params_parsed)
```
