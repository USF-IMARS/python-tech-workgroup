```python
# === load a pickle from a URL
# ====================================================================================================
# This an http URL of a pickle file we want to open. It can come from anywhere.
PICKLE_URL = "https://github.com/USF-IMARS/bb3_matchup/blob/main/data/bb3_df_first_cruise.pickle?raw=true"

import urllib.request
import pickle

response = urllib.request.urlopen(PICKLE_URL)

df = pickle.load(response)  

# display(df.head())

```

```python
import matplotlib.pyplot

# from: https://numpy.org/doc/stable/reference/generated/numpy.histogram.html
_ = matplotlib.pyplot.hist(df['470nm_data'], bins='auto')  # arguments are passed to np.histogram
matplotlib.pyplot.title("Histogram with 'auto' bins")

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(20,5)
matplotlib.pyplot.xlim([0,2000])
matplotlib.pyplot.show()

```

```python
# histogram w/ plotly
import plotly.express as px
df = px.data.tips()
fig = px.histogram(df, x="total_bill")
fig.show()
```
