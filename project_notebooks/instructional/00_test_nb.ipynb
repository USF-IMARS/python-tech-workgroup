```python
# python comments start with `#`
#
# In this chunk we define a function.
# Code in the function will not be run until
# the function is called like `my_func(number=1)`

def my_func(number=0):
    """
    Three quotes can be used to create a long comment.
    These triple-quoted things are called docstrings.
    Docstrings do more than just comment though.
    
    This function will print the given number.
    """
    print(
        '=== Welcome to my func v0.0.1 ===\n'
        '================================='
    )
    
    # example use of format strings:
    print(f'The number is {number}\n')
    # NOTE: the \n adds an empty line at the end.
    #       \n is for "newline".
```

```python
# In this chunk we call the function that was previously defined.
# Make sure you ran the chunk above first or you will get an 
# error because the function was not defined before we attempt
# to call the function here.

print('first with no args:')
my_func()

print('now we use the arg:')
my_func(number=123)

print('the arg does not really have to be a number though:')
my_func(number='this is not a number')
```
